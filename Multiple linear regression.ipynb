# Multiple linear regression 

Multiple linear regression uses two or more independent variables to predict the values of the dependent variable.

### Dependent Variable

The dependent variable is what you measure in the experiment and what is affected during the experiment. It's dependent on the independent variable.

If you are studying the effect of different amounts of sunlight on plant growth, the height of the plants is the dependent variable because it depends on the amount of sunlight the plants receive.

Often denoted as Y in equations, it is the variable that researchers are most interested in understanding, explaining, or predicting.

### Independent Variable

The independent variable is what is varied during the experiment. It is what the investigator thinks will affect the dependent variable.

In the same experiment of plant growth, the amount of sunlight would be the independent variable because it is being manipulated to observe its effect on the plants.

Often denoted as X in equations, it is used as a tool to understand, explain, or predict changes in the dependent variable.

In the Context of linear regression the dependent variable is what you are trying to predict or explain. It is plotted on the Y-axis. The independent variable is used to make predictions on the dependent variable and is plotted on the X-axis.
